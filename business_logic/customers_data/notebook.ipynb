# Notebook d'exploration sur une population d'humains

Jeu de données sur les humains

```python
humans = [
    {"name": "Alice", "age": 30, "city": "New York"},
    {"name": "Bob", "age": 25, "city": "Los Angeles"},
    {"name": "Charlie", "age": 35, "city": "Chicago"},  
]

humans 
```

Je filtre les adultes de plus de 30 ans avec une compréhension de liste :

```python
[h for h in humans if h["age"] > 30]
```

```python
humans_filtered = []
for h in humans:
    if h["age"] > 30:
        humans_filtered.append(h)
```

Je filtre les adultes de plus de 30 ans avec un filter:

```python
def is_adult(human) -> bool:
    return human["age"] > 20
```

```python
list(filter(is_adult, humans))
```

```python
def add_adult_status(human):
    # human["is_adult"] = is_adult(human)
    # return human
    return {**human, "is_adult": is_adult(human)}
    
```

```python
new_adults = list(map(add_adult_status, humans))
new_adults
```

```python
def add_human_category(human):
    return {**human, "category": "adult" if human["is_adult"] else "minor"}

'''
if human["is_adult"]:
    human["category"] = "adult"
else:
    human["category"] = "minor"
'''
```

```python
list(map(add_human_category, new_adults))
```
